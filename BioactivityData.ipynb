## **Import New Client Function**

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

##  **Search Target by Gene Name**

```python
target = new_client.target
target_query = target.search('GRB2')
targets = pd.DataFrame.from_dict(target_query)
targets
```

## **Retrieve ChEMBL ID of one of the entries**

Assign the first entry to the **selected_target** variable.

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

## **Retrieve Bioactivity Data for GRB2-Mouse**

[More information about bioactivity data](https://chembl.gitbook.io/chembl-interface-documentation/frequently-asked-questions/chembl-data-questions#can-you-provide-more-details-on-bioactivity-data-standardisation)

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target)
```

```python
df = pd.DataFrame.from_dict(res)
```

## **Coronavirus!**

You can also search by common name.

```python
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target)
```

```python
df = pd.DataFrame.from_dict(res)
df
```
